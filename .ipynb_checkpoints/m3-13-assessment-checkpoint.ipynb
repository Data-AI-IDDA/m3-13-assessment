## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
from scipy.stats import pearsonr, spearmanr

pairs = [
    ("mpg", "weight"),
    ("horsepower", "displacement"),
    ("mpg", "horsepower")
]

results = []

for x, y in pairs:
    pearson_corr, pearson_p = pearsonr(mpg[x], mpg[y])
    spearman_corr, spearman_p = spearmanr(mpg[x], mpg[y])

    results.append({
        "pair": f"{x} vs {y}",
        "pearson_corr": pearson_corr,
        "pearson_p": pearson_p,
        "spearman_corr": spearman_corr,
        "spearman_p": spearman_p
    })

corr_df = pd.DataFrame(results)
corr_df
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = mpg.select_dtypes(include=np.number).corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
for x, y in pairs:
    plt.figure(figsize=(6, 4))
    
    sns.regplot(data=mpg, x=x, y=y, scatter_kws={"alpha":0.6})
    
    r, _ = pearsonr(mpg[x], mpg[y])
    
    plt.title(f"{x} vs {y}")
    plt.text(
        0.05, 0.95,
        f"Pearson r = {r:.2f}",
        transform=plt.gca().transAxes,
        fontsize=12,
        verticalalignment='top'
    )
    
    plt.show()
```

### Interpretation

*When is Pearson vs. Spearman more appropriate? Use your results as examples.*

Pearson and Spearman correlations are both used to measure relationships between variables, but they are appropriate in different situations.

Pearson correlation measures linear relationships, meaning it works best when the data follows a straight-line pattern. In contrast, Spearman correlation is based on ranks and measures monotonic relationships, so it can capture patterns that are not strictly linear but still consistently increasing or decreasing.

In my results, the relationship between *mpg* and *weight* shows a strong negative correlation in both Pearson (-0.83) and Spearman (-0.88). This suggests that the relationship is not only strong but also fairly consistent and monotonic. Similarly, *horsepower* and *displacement* have very high positive correlations in both methods (around 0.90 and 0.88), indicating a strong linear relationship.

However, in all pairs, the Spearman correlation is slightly stronger than Pearson. This may suggest that the relationships are not perfectly linear and contain some non-linear patterns or outliers. In such cases, Spearman can be more robust and reliable.

All p-values are extremely small (close to 0), which indicates that these correlations are statistically significant and unlikely to have occurred by chance.

In conclusion, Pearson is more appropriate when the relationship is linear, while Spearman is preferred when the relationship is monotonic or when the data may contain outliers or non-linear patterns.

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
from scipy.stats import shapiro, levene, f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

plt.figure(figsize=(8, 5))
sns.boxplot(data=mpg, x="origin", y="mpg")
plt.title("MPG Distribution by Origin")
plt.show()
```

```python
groups = mpg["origin"].unique()

for group in groups:
    data = mpg[mpg["origin"] == group]["mpg"]
    stat, p = shapiro(data)
    print(f"{group} → Shapiro p-value: {p:.4f}")
```

```python
usa = mpg[mpg["origin"] == "usa"]["mpg"]
europe = mpg[mpg["origin"] == "europe"]["mpg"]
japan = mpg[mpg["origin"] == "japan"]["mpg"]

stat, p = levene(usa, europe, japan)
print(f"Levene p-value: {p:.4f}")
```

```python
f_stat, p_value = f_oneway(usa, europe, japan)

print(f"F-statistic: {f_stat:.4f}")
print(f"p-value: {p_value:.4e}")
```

```python
from scipy.stats import kruskal

# I use also Kruskal-Wallis because use's and europe's sharpio p-value is smaller than 0.05
stat_kw, p_kw = kruskal(usa, europe, japan)

print(f"Kruskal-Wallis H-statistic: {stat_kw:.4f}")
print(f"Kruskal-Wallis p-value: {p_kw:.4e}")
```

```python
grand_mean = mpg["mpg"].mean()

ss_between = sum([
    len(group) * (group.mean() - grand_mean)**2
    for group in [usa, europe, japan]
])

ss_total = sum((mpg["mpg"] - grand_mean)**2)

eta_squared = ss_between / ss_total

print(f"Eta-squared: {eta_squared:.4f}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=mpg["mpg"], groups=mpg["origin"], alpha=0.05)
print(tukey)
```

### Interpretation

*Which groups differ? What does the effect size tell you?*

First, a quick note on our methodology: During our initial assumption checks, we noticed that the data didn't perfectly align with standard ANOVA rules — specifically, fuel efficiency for the USA and Europe wasn't perfectly normally distributed, and the variances across groups were unequal. To be completely safe and mathematically rigorous, we ran a non-parametric Kruskal-Wallis test as a backup. Since it strongly confirmed our ANOVA results, we can be highly confident in our findings!

The results show that there are significant differences between all origin groups. While the difference between Europe and Japan is relatively smaller, it is still statistically significant. In contrast, the differences between the USA and the other two groups (Europe and Japan) are much larger.

The high eta-squared value (0.33) indicates that these differences are not only statistically significant but also practically important, meaning that the origin of the car has a strong effect on fuel efficiency (mpg).

In conclusion, although Europe and Japan are closer to each other, the USA stands out as being significantly different from both, with noticeably lower fuel efficiency.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols].to_numpy()
y = mpg["origin"].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
# Train classifiers and generate predictions
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

log_pred = log_model.predict(X_test)

tree_model = DecisionTreeClassifier(random_state=SEED)
tree_model.fit(X_train, y_train)

tree_pred = tree_model.predict(X_test)
```

```python
# Classification reports
print("=== Logistic Regression ===")
print(classification_report(y_test, log_pred))


print("=== Decision Tree ===")
print(classification_report(y_test, tree_pred))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """Compute bootstrap 95% confidence interval for a classification metric.

    Returns
    -------
    point_estimate : float
    ci_lower : float
    ci_upper : float
    boot_scores : np.ndarray
    """
    np.random.seed(seed)
    scores = []
    n = len(y_true)
    
    for _ in range(n_boot):
        idx = np.random.choice(range(n), size=n, replace=True)
        
        if len(np.unique(y_true[idx])) < 2:
            continue
        
        score = metric_fn(y_true[idx], y_pred[idx], average="macro")
        scores.append(score)
    
    lower = np.percentile(scores, 2.5)
    upper = np.percentile(scores, 97.5)
    return np.mean(scores), lower, upper
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)
log_f1 = bootstrap_metric(y_test, log_pred, f1_score)
log_precision = bootstrap_metric(y_test, log_pred, precision_score)
log_recall = bootstrap_metric(y_test, log_pred, recall_score)

print("Logistic Regression CI:")
print("F1:", log_f1)
print("Precision:", log_precision)
print("Recall:", log_recall)

print("----------------------------")
print("----------------------------")
print("----------------------------")
print("----------------------------")

tree_f1 = bootstrap_metric(y_test, tree_pred, f1_score)
tree_precision = bootstrap_metric(y_test, tree_pred, precision_score)
tree_recall = bootstrap_metric(y_test, tree_pred, recall_score)

print("Decision Tree CI:")
print("F1:", tree_f1)
print("Precision:", tree_precision)
print("Recall:", tree_recall)
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)
models = ["Logistic", "Tree"]
metrics = ["F1", "Precision", "Recall"]

# mean values
log_means = [log_f1[0], log_precision[0], log_recall[0]]
tree_means = [tree_f1[0], tree_precision[0], tree_recall[0]]

x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x - width/2, log_means, width, label="Logistic Regression")
plt.bar(x + width/2, tree_means, width, label="Decision Tree")

plt.xticks(x, metrics)
plt.ylabel("Score")
plt.title("Model Comparison (Mean Metrics)")
plt.legend()

plt.show()
```

### Recommendation

*Which model would you recommend? How confident are you based on the CIs?*

I would recommend the Decision Tree model because it achieves higher F1, precision, and recall across all classes. Its bootstrap confidence intervals are slightly narrower and shifted higher compared to Logistic Regression, indicating both better performance and comparable stability. Therefore, the improvement is consistent and not due to random variation in the test set.
