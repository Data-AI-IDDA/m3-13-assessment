## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
# Correlation coefficients and p-values


df = mpg[["mpg", "weight", "horsepower", "displacement", "acceleration"]]

pairs = [("mpg","weight"), ("horsepower","displacement"), ("acceleration","mpg")]

for x, y in pairs:
    pearson_r, pearson_p = stats.pearsonr(df[x], df[y])
    spearman_r, spearman_p = stats.spearmanr(df[x], df[y])
    print(f"\n{x} vs {y}")
    print(f"Pearson r = {pearson_r:.3f}, p = {pearson_p:.4g}")
    print(f"Spearman ρ = {spearman_r:.3f}, p = {spearman_p:.4g}")
```

```python
# Correlation heatmap

# Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(method="pearson"), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Pearson)")
plt.show()
```

```python
# Scatter plots with regression lines

# Scatter plots with regression lines
def scatter_reg(x, y):
    r, _ = stats.pearsonr(df[x], df[y])
    sns.lmplot(x=x, y=y, data=df, height=4)
    plt.title(f"{x} vs {y} (r = {r:.2f})")
    plt.show()

scatter_reg("weight", "mpg")
scatter_reg("horsepower", "displacement")
```

### Interpretation

*When is Pearson vs. Spearman more appropriate? Use your results as examples.*

Your answer here…

Pearson correlation measures linear relationships, while Spearman correlation measures monotonic relationships (relationships that increase or decrease consistently but may not be linear).

In this dataset, weight vs. mpg shows a strong negative correlation, indicating heavier vehicles tend to have lower fuel efficiency. Both Pearson and Spearman correlations are similarly strong, showing the relationship is monotonic and approximately linear.

Similarly, horsepower vs. displacement shows a strong positive relationship, meaning larger engines tend to produce more power.

Pearson is appropriate when relationships are linear and normally distributed, while Spearman is preferable when data contain outliers or non-linear monotonic patterns.

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization by origin

# Distribution visualization by origin
sns.boxplot(x="origin", y="mpg", data=mpg)
plt.title("MPG Distribution by Origin")
plt.show()
```

```python
# Assumption checks: Shapiro-Wilk + Levene's test

# Assumption checks: Shapiro-Wilk + Levene's test
groups = [mpg[mpg["origin"]==o]["mpg"] for o in mpg["origin"].unique()]

print("Shapiro-Wilk Normality Tests:")
for name, g in zip(mpg["origin"].unique(), groups):
    print(name, stats.shapiro(g))

print("\nLevene’s Test for Equal Variance:")
print(stats.levene(*groups))
```

```python
# One-way ANOVA

# One-way ANOVA
f_stat, p_val = stats.f_oneway(*groups)
print("ANOVA F-statistic:", f_stat)
print("ANOVA p-value:", p_val)

# Effect size: Eta squared
grand_mean = mpg["mpg"].mean()
ss_between = sum(len(g)*(g.mean()-grand_mean)**2 for g in groups)
ss_total = sum((mpg["mpg"]-grand_mean)**2)
eta_sq = ss_between/ss_total
print("Eta squared:", eta_sq)
```

```python
# Post-hoc test (Tukey's HSD)

# Post-hoc test (Tukey's HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(mpg["mpg"], mpg["origin"])
print(tukey)
```

### Interpretation

*Which groups differ? What does the effect size tell you?*

Your answer here…

The ANOVA test indicates that mean MPG differs significantly across regions (p < 0.05). Tukey’s HSD results show that cars from different regions are significantly different in terms of fuel efficiency.

The effect size (eta-squared) quantifies the magnitude of the difference. A larger value indicates stronger group differences, meaning that vehicle origin explains a substantial portion of MPG variation.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

SEED = 42

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols]
y = mpg["origin"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
# Train classifiers and generate predictions

log_reg = make_pipeline(StandardScaler(), LogisticRegression(max_iter=500))
tree = DecisionTreeClassifier(random_state=SEED)

log_reg.fit(X_train, y_train)
tree.fit(X_train, y_train)

pred_lr = log_reg.predict(X_test)
pred_tree = tree.predict(X_test)
```

```python
# Classification reports

print("Logistic Regression:")
print(classification_report(y_test, pred_lr))

print("Decision Tree:")
print(classification_report(y_test, pred_tree))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """Compute bootstrap 95% confidence interval for a classification metric."""
    
    rng = np.random.default_rng(seed)
    scores = []

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    for _ in range(n_boot):
        idx = rng.integers(0, len(y_true), len(y_true))
        scores.append(metric_fn(y_true[idx], y_pred[idx], average="weighted"))

    lower = np.percentile(scores, 2.5)
    upper = np.percentile(scores, 97.5)

    return np.mean(scores), lower, upper, np.array(scores)
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)

metrics = {"F1": f1_score, "Precision": precision_score, "Recall": recall_score}
results = []

for name, preds in [("LogReg", pred_lr), ("Tree", pred_tree)]:
    for mname, fn in metrics.items():
        mean, low, high, _ = bootstrap_metric(y_test, preds, fn)
        print(f"{name} {mname}: {mean:.3f}, CI=({low:.3f}, {high:.3f})")
        results.append((name, mname, mean, low, high))
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)

results_df = pd.DataFrame(results, columns=["Model","Metric","Mean","Low","High"])

plt.figure(figsize=(7,5))
for i, row in results_df.iterrows():
    plt.plot([row["Low"], row["High"]], [i,i], marker="o")
plt.yticks(range(len(results_df)), results_df["Model"]+"-"+results_df["Metric"])
plt.xlabel("Score")
plt.title("Bootstrap Confidence Intervals")
plt.show()
```

### Recommendation

*Which model would you recommend? How confident are you based on the CIs?*

Your answer here…


The recommended model is the one with consistently higher scores and tighter confidence intervals. If the confidence intervals overlap significantly between models, their performance is statistically similar. Choose the model that balances accuracy with stability across metrics.
