## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
# Correlation coefficients and p-values

from scipy.stats import pearsonr, spearmanr

pairs = [
    ("mpg", "horsepower"),
    ("mpg", "weight"),
    ("horsepower", "acceleration"),
]

for x, y in pairs:
    pearson_r, pearson_p = pearsonr(mpg[x], mpg[y])
    spearman_r, spearman_p = spearmanr(mpg[x], mpg[y])
    print(f"{x} vs {y}")
    print(f"  Pearson r = {pearson_r:.3f}, p = {pearson_p:.3e}")
    print(f"  Spearman r = {spearman_r:.3f}, p = {spearman_p:.3e}\n")
```

## Pair‑wise Comparison

### mpg vs horsepower
- Pearson \(r = -0.778\), Spearman \(ρ = -0.854\)  
- Reported p‑values: \(7.032e-81), \(1.619e-112)\)  
- **Comparison:** Both p‑values are far below 0.05 → Significant correlation.

### mpg vs weight
- Pearson \(r = -0.832\), Spearman \(ρ = -0.876\)  
- Reported p‑values: \(6.015e-102), \(2.662e-125)  
- **Comparison:** Both p‑values are far below 0.05 → Significant correlation.

### horsepower vs acceleration
- Pearson \(r = -0.689\), Spearman \(ρ = -0.658\)  
- Reported p‑values: \(1.582e-56), \(5.158e-50)  
- **Comparison:** Both p‑values are far below 0.05 → Significant correlation.

## Pearson vs Spearman Correlation — When to Use (with Examples)

### 1. Pearson Correlation (r)
Use Pearson when:
- Relationship is **linear**
- Variables are **continuous**
- Data is approximately **normally distributed**
- No strong **outliers**
- You care about **actual values and distances**

→ Measures **linear association**

---

### 2. Spearman Correlation (ρ)
Use Spearman when:
- Relationship is **monotonic** (consistently increasing/decreasing)
- Relationship may be **nonlinear**
- Data is **non-normal**
- There are **outliers**
- You care about **rank/order**, not exact values

→ Measures **rank-based (monotonic) association**

---

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization by origin


# Boxplot: MPG distribution by origin
plt.figure(figsize=(8,6))
sns.boxplot(data=mpg, x="origin", y="mpg", hue="origin", palette="Set2", legend=False)
plt.title("MPG Distribution by Origin (Boxplot)")
plt.show()

# Violin plot: MPG distribution by origin
plt.figure(figsize=(8,6))
sns.violinplot(data=mpg, x="origin", y="mpg", hue="origin", palette="Set2", inner="quartile", legend=False)
plt.title("MPG Distribution by Origin (Violin Plot)")
plt.show()
```

```python
# Assumption checks: Shapiro-Wilk + Levene's test

from scipy.stats import shapiro

for group in mpg["origin"].unique():
    stat, p = shapiro(mpg.loc[mpg["origin"] == group, "mpg"])
    print(f"{group}: W = {stat:.3f}, p = {p:.3e}")


from scipy.stats import levene

usa = mpg.loc[mpg["origin"] == "usa", "mpg"]
japan = mpg.loc[mpg["origin"] == "japan", "mpg"]
europe = mpg.loc[mpg["origin"] == "europe", "mpg"]

stat, p = levene(usa, japan, europe)
print(f"Levene’s test: W = {stat:.3f}, p = {p:.3e}")


```

### ANOVA Assumption Checks

### Shapiro–Wilk (Normality)
- USA: W = 0.934, p = 4.76e-09 → Not normal
- Japan: W = 0.980, p = 0.242 → Normal
- Europe: W = 0.946, p = 0.0054 → Not normal

**Conclusion:** Normality violated for USA and Europe.

### Levene’s Test (Equal Variances)
- W = 0.086, p = 0.918 → Variances equal

**Conclusion:** Homogeneity of variance holds.


```python
# One-way ANOVA
F_stat, p_value = stats.f_oneway(usa, japan, europe)

# Eta-squared calculation
grand_mean = mpg["mpg"].mean()
ss_between = sum(len(group) * (group.mean() - grand_mean)**2 
                 for group in [usa, japan, europe])
ss_total = sum((mpg["mpg"] - grand_mean)**2)

eta_sq = ss_between / ss_total

print(f"F-statistic: {F_stat:.3f}")
print(f"p-value: {p_value:.3e}")
print(f"Eta-squared: {eta_sq:.3f}")
```

### One‑Way ANOVA (mpg by origin)

- F-statistic = 96.602  
- p-value = 8.674e-35  
- η² = 0.332  

**Interpretation:**  
Mean mpg differs significantly across origins.  
Eta‑squared (0.332) indicates a large effect size — origin explains ~33% of the variance in mpg.


```python
# Kruskal–Wallis test
H_stat, p_value = stats.kruskal(usa, japan, europe)

print(f"H-statistic: {H_stat:.3f}")
print(f"p-value: {p_value:.3e}")
```

### Kruskal–Wallis Test (more appropriate)
- H-statistic = 132.087  
- p-value = 2.078e-29  
**Interpretation:** mpg distributions differ significantly across origins.  
Since normality was violated, Kruskal–Wallis provides the more appropriate inference.

```python
# Post-hoc test (Tukey's HSD)

from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Tukey's HSD post-hoc test
tukey = pairwise_tukeyhsd(endog=mpg["mpg"], groups=mpg["origin"], alpha=0.05)
print(tukey)
```

### Interpretation

## Tukey’s HSD (Post‑Hoc Results)

### Significant Group Differences
- **Europe vs Japan** → Significant (Japan > Europe)

- **Japan vs USA** → Significant (Japan > USA)  

**Interpretation:**  
All three origin groups differ significantly in mpg.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols].to_numpy()
y = np.array(mpg["origin"])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
mpg
```

```python
# Train classifiers and generate predictions

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# 1. Initialize Random Forest
# Ensembles like RF are robust to different scales (e.g., weight vs. acceleration)
rf_model = RandomForestClassifier(random_state=SEED)

# 2. Initialize Logistic Regression
# IMPORTANT: Logistic Regression is sensitive to scale. 
# We use a Pipeline to scale the features (mpg, weight, etc.) automatically.
lr_model = make_pipeline(
    StandardScaler(), 
    LogisticRegression(random_state=SEED)
)

# 3. Train both models
rf_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)

# 4. Generate predictions
rf_preds = rf_model.predict(X_test)
lr_preds = lr_model.predict(X_test)

print("Classifiers trained: Random Forest & Logistic Regression (scaled).")
```

```python
# Classification reports

# 1. Generate reports for Random Forest
print("=== RANDOM FOREST CLASSIFICATION REPORT ===")
rf_report = classification_report(y_test, rf_preds)
print(rf_report)

print("-" * 50)

# 2. Generate reports for Logistic Regression (Pipeline)
print("=== LOGISTIC REGRESSION CLASSIFICATION REPORT ===")
lr_report = classification_report(y_test, lr_preds)
print(lr_report)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    rng = np.random.default_rng(seed)
    boot_scores = []
    
    # 1. Point Estimate (the original score)
    point_estimate = metric_fn(y_true, y_pred, average='macro')
    
    # 2. Resampling loop
    indices = np.arange(len(y_true))
    for _ in range(n_boot):
        # Sample indices with replacement
        resample_idx = rng.choice(indices, size=len(indices), replace=True)
        
        # Calculate metric on this specific resample
        score = metric_fn(y_true[resample_idx], y_pred[resample_idx], average='macro')
        boot_scores.append(score)
        
    boot_scores = np.array(boot_scores)
    
    # 3. Compute Percentiles (2.5th and 97.5th for 95% CI)
    ci_lower = np.percentile(boot_scores, 2.5)
    ci_upper = np.percentile(boot_scores, 97.5)
    
    return point_estimate, ci_lower, ci_upper, boot_scores
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)


# Define the metrics we want to analyze
metrics_to_run = {
    "F1 (Macro)": f1_score,
    "Precision (Macro)": precision_score,
    "Recall (Macro)": recall_score
}

# Define the models and their corresponding predictions
model_predictions = {
    "Random Forest": rf_preds,
    "Logistic Regression": lr_preds
}

# Results loop
for model_name, preds in model_predictions.items():
    print(f"\n--- {model_name} Bootstrapped Metrics (n=2000) ---")
    print(f"{'Metric':<15} | {'Estimate':<10} | {'95% CI':<20}")
    print("-" * 50)
    
    for metric_name, metric_fn in metrics_to_run.items():
        # Using y_test and the specific model's predictions
        est, low, high, _ = bootstrap_metric(y_test, preds, metric_fn)
        
        print(f"{metric_name:<15} | {est:.3f}    | [{low:.3f}, {high:.3f}]")
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)

# 1. Store your results in a DataFrame for easy plotting
results_list = []
model_preds = {"Random Forest": rf_preds, "Logistic Regression": lr_preds}
metrics_map = {"F1 (Macro)": f1_score, "Precision (Macro)": precision_score, "Recall (Macro)": recall_score}

for model_name, preds in model_preds.items():
    for metric_name, metric_fn in metrics_map.items():
        est, low, high, _ = bootstrap_metric(y_test, preds, metric_fn)
        results_list.append({
            "Model": model_name, 
            "Metric": metric_name, 
            "Estimate": est, 
            "CI_Lower": low, 
            "CI_Upper": high
        })

df_res = pd.DataFrame(results_list)

# 2. Plotting
plt.figure(figsize=(10, 6))
colors = ['#2c3e50', '#e74c3c'] # Navy for RF, Red for LR
metrics = df_res['Metric'].unique()

for i, model in enumerate(df_res['Model'].unique()):
    subset = df_res[df_res['Model'] == model]
    
    # Calculate error lengths for the whiskers
    y_err = [subset['Estimate'] - subset['CI_Lower'], subset['CI_Upper'] - subset['Estimate']]
    
    # Offset x-axis so the dots don't overlap
    x_pos = np.arange(len(metrics)) + (i * 0.2 - 0.1)
    
    plt.errorbar(x_pos, subset['Estimate'], yerr=y_err, fmt='o', 
                 capsize=8, label=model, color=colors[i], markersize=10)

plt.xticks(np.arange(len(metrics)), metrics)
plt.ylabel('Score (Macro Average)')
plt.title('95% Bootstrap Confidence Intervals: RF vs. Logistic Regression')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.legend()
plt.show()
```
