```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

SEED = 42
np.random.seed(SEED)

mpg = sns.load_dataset("mpg").dropna()

print(f"Rows: {len(mpg)}, Columns: {mpg.shape[1]}")
print(f"Origin groups: {mpg['origin'].unique().tolist()}")
mpg.head()
```

## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
# Correlation coefficients and p-values
pairs = [('mpg', 'weight'), ('displacement', 'horsepower'), ('horsepower', 'weight')]

results = []
for col1, col2 in pairs:
    pearson_r, p_p = stats.pearsonr(mpg[col1], mpg[col2])
    spearman_r, p_s = stats.spearmanr(mpg[col1], mpg[col2])
    results.append({
        'Pair': f"{col1} vs {col2}",
        'Pearson r': pearson_r, 'Pearson p': p_p,
        'Spearman r': spearman_r, 'Spearman p': p_s
    })

results_df = pd.DataFrame(results)
print(results_df.to_string(index=False))
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 7))
numeric_df = mpg.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Numerical Variables Correlation Heatmap")
plt.show()
```

```python
# Scatter plots with regression lines
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: mpg vs weight
sns.regplot(data=mpg, x='weight', y='mpg', ax=axes[0], 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[0].set_title(f"MPG vs Weight\nPearson: {mpg['mpg'].corr(mpg['weight'], method='pearson'):.2f}")

# Plot 2: displacement vs horsepower
sns.regplot(data=mpg, x='displacement', y='horsepower', ax=axes[1], 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[1].set_title(f"Displacement vs Horsepower\nPearson: {mpg['displacement'].corr(mpg['horsepower'], method='pearson'):.2f}")

plt.tight_layout()
plt.show()
```

### Interpretation

*When is Pearson vs. Spearman more appropriate? Use your results as examples.*
Pearson correlation measures the relationship between two variables. For example, for power and horsepower in relation to displacement, Pearson is fully established here.

Spearman measures the monotonic relationship between variables ,when one increases, the other also increases or decreases, but not necessarily in a straight line. If we look at the relationship between mpg and weight, fuel efficiency decreases non-linearly as weight increases. Spearman is more reliable in this type of non-linear but regular relationship or when there are outliers in the data.

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization by origin
plt.figure(figsize=(12, 6))

sns.violinplot(data=mpg, x='origin', y='mpg',hue='origin', inner="quart", palette="muted")
sns.swarmplot(data=mpg, x='origin', y='mpg', color="white", alpha=0.4, size=3) 

plt.title("Distribution of MPG by Origin", fontsize=14)
plt.xlabel("Origin (Region)", fontsize=12)
plt.ylabel("Miles Per Gallon (MPG)", fontsize=12)
plt.show()
```

```python
# Assumption checks: Shapiro-Wilk + Levene's test
# We separate the groups
usa = mpg[mpg['origin'] == 'usa']['mpg']
japan = mpg[mpg['origin'] == 'japan']['mpg']
europe = mpg[mpg['origin'] == 'europe']['mpg']

print("--- 1. Normality Test (Shapiro-Wilk) ---")
# We check whether each group is normally distributed or not
for name, group in zip(['USA', 'Japan', 'Europe'], [usa, japan, europe]):
    stat, p = stats.shapiro(group)
    status = "Normal" if p > 0.05 else "Not Normal"
    print(f"{name}: p-value = {p:.4f} ({status})")

print("\n--- 2. Equality of Variances (Levene Testi) ---")
# We check the equality of variance between groups
stat_l, p_l = stats.levene(usa, japan, europe)
status_l = "Equal" if p_l > 0.05 else "Not Equal"
print(f"Levene p-value = {p_l:.4f} ({status_l})")

plt.figure(figsize=(15, 5))
for i, (name, group) in enumerate(zip(['USA', 'Japan', 'Europe'], [usa, japan, europe]), 1):
    plt.subplot(1, 3, i)
    stats.probplot(group, dist="norm", plot=plt)
    plt.title(f"Q-Q Plot: {name}")
plt.tight_layout()
plt.show()
```

```python
# One-way ANOVA
f_stat, p_val = stats.f_oneway(usa, japan, europe)

# Formula: SS_between / SS_total
all_mpg = mpg['mpg']
ss_total = np.sum((all_mpg - all_mpg.mean())**2)
ss_between = (len(usa)*(usa.mean() - all_mpg.mean())**2 + 
              len(japan)*(japan.mean() - all_mpg.mean())**2 + 
              len(europe)*(europe.mean() - all_mpg.mean())**2)
eta_sq = ss_between / ss_total

print(f"F-statistic: {f_stat:.2f}")
print(f"p-value: {p_val:.2e}")
print(f"Eta-squared: {eta_sq:.3f}")
```

```python
# Post-hoc test (Tukey's HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=mpg['mpg'], groups=mpg['origin'], alpha=0.05)
print(tukey)

fig = tukey.plot_simultaneous(xlabel="MPG")
plt.title("Tukey HSD Comparison")
plt.show()
```

### Interpretation

*Which groups differ? What does the effect size tell you?*

The Shapiro-Wilk test shows that some groups may not be normally distributed (p < 0.05). However, ANOVA is still reliable when the sample size is large. The Levene test shows that the variances are not equal, so MPG values change differently across groups.

The ANOVA result has a very small p-value (p < 0.001), so we reject the null hypothesis. This means the average MPG is not the same for all regions. Eta-squared shows that the car’s origin explains about X% of the MPG differences.

The Tukey HSD test shows clear differences between groups. In particular, cars from the USA are significantly different from Japan and Europe. Cars from the USA have lower MPG, which means they use more fuel.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols].values
y = mpg["origin"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
# Train classifiers and generate predictions
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# We are crearting models
# StandardScaler is important for Logistic Regression,because it works based on distance
log_reg = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000, random_state=SEED))
dtree = DecisionTreeClassifier(random_state=SEED)

# We train the models
log_reg.fit(X_train, y_train)
dtree.fit(X_train, y_train)

# We get predictions over the test set
y_pred_log = log_reg.predict(X_test)
y_pred_dt = dtree.predict(X_test)
```

```python
# Classification reports
print("--- Logistic Regression Performance ---")
print(classification_report(y_test, y_pred_log))

print("\n--- Decision Tree Performance ---")
print(classification_report(y_test, y_pred_dt))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """Compute bootstrap 95% confidence interval for a classification metric.

    Returns
    -------
    point_estimate : float
    ci_lower : float
    ci_upper : float
    boot_scores : np.ndarray
    """
    # Your code here
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    rng = np.random.default_rng(seed)
    boot_scores = []
    indices = np.arange(len(y_true))
    
    for _ in range(n_boot):
        resample_idx = rng.choice(indices, size=len(indices), replace=True)
        resample_true = y_true[resample_idx]
        resample_pred = y_pred[resample_idx]
        score = metric_fn(resample_true, resample_pred, average='weighted')
        boot_scores.append(score)
    
    boot_scores = np.array(boot_scores)
    point_estimate = metric_fn(y_true, y_pred, average='weighted')
    ci_lower = np.percentile(boot_scores, 2.5)
    ci_upper = np.percentile(boot_scores, 97.5)
    
    return point_estimate, ci_lower, ci_upper, boot_scores
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)
metrics_list = {'F1': f1_score, 'Precision': precision_score, 'Recall': recall_score}
models_preds = {'Logistic Regression': y_pred_log, 'Decision Tree': y_pred_dt}

boot_results = []

for model_name, preds in models_preds.items():
    for metric_name, fn in metrics_list.items():
        point, low, high, _ = bootstrap_metric(y_test, preds, fn)
        boot_results.append({
            'Model': model_name,
            'Metric': metric_name,
            'Estimate': point,
            'CI_Lower': low,
            'CI_Upper': high
        })

boot_df = pd.DataFrame(boot_results)
print(boot_df)
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)
plt.figure(figsize=(14, 5))

for i, metric in enumerate(['F1', 'Precision', 'Recall']):
    plt.subplot(1, 3, i+1)
    data = boot_df[boot_df['Metric'] == metric]
    
    for j, row in data.iterrows():
        plt.errorbar(x=row['Estimate'], y=row['Model'], 
                     xerr=[[row['Estimate'] - row['CI_Lower']], [row['CI_Upper'] - row['Estimate']]],
                     fmt='o', capsize=6, elinewidth=2, markeredgewidth=2)
    
    plt.title(f"95% Bootstrap CI for {metric}")
    plt.xlabel("Score")
    if i > 0: plt.yticks([])

plt.tight_layout()
plt.show()
```

### Recommendation

*Which model would you recommend? How confident are you based on the CIs?*

    The confidence intervals of the Logistic Regression model are narrower and higher than those of the Decision Tree. This means Logistic Regression is more stable on new data.
    
    The Decision Tree has wider intervals, so it has more uncertainty. Therefore, Logistic Regression is more reliable for predicting the origin in this dataset.
