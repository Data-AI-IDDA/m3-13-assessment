## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
# Correlation coefficients and p-values
# Correlation coefficients and p-values

# Pearson korrelyasiya koeffisiyentlerini ve p-value hesablayiram 
#parametrleri soldan saga pairledim
p_r_m, p_p_m = stats.pearsonr(mpg['mpg'], mpg['weight'])
p_r_d, p_p_d = stats.pearsonr(mpg['displacement'], mpg['horsepower'])
p_r_c, p_p_c = stats.pearsonr(mpg['cylinders'], mpg['horsepower'])

# Hemin cutlerin Spearman korrelyasiya koeffisiyentlerini və p-value hesablayiram
s_r_m, s_p_m = stats.spearmanr(mpg['mpg'], mpg['weight'])
s_r_d, s_p_d = stats.spearmanr(mpg['displacement'], mpg['horsepower'])
s_r_c, s_p_c = stats.spearmanr(mpg['cylinders'], mpg['horsepower'])

# dataframe seklinde bir yere yigiram dictionary yaradilir
result = {
    'Variable Pair': ['mpg vs weight', 'displacement vs horsepower', 'cylinders vs displacement'],
    
    # Pearson korrelyasiya neticeleri
    'Pearson R': [p_r_m, p_r_m, p_r_c],
    'Pearson p-value': [p_p_m, p_p_d, p_p_c],
    
    # Spearman korrelyasiya neticeleri
    'Spearman R': [s_r_m, s_r_d, s_r_c],
    'Spearman p-value': [s_p_m, s_p_d, s_p_c]
}

# Dictionary DataFrame-ə çevrilir
corr_df = pd.DataFrame(result)

# Nəticə cədvəli göstərilir
corr_df
```

```python
# Correlation heatmap
sns.heatmap(mpg.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Scatter plots with regression lines
#oxuna bilen plot yaradiram diapozon gorsensin deye
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.regplot(data=mpg, x='displacement', y='weight', ax=axes[0])
axes[0].set_title('Displacement vs Weight')
sns.regplot(data=mpg, x='weight', y='cylinders', ax=axes[1])
axes[1].set_title('Weight vs Cylinders')

plt.tight_layout()
plt.show()
```

### Interpretation

*When is Pearson vs. Spearman more appropriate? Use your results as examples.*

The Spearman correlations are generally higher than the Pearson correlations, indicating that the relationships among the variables are highly consistent even if they are not strictly linear. Since all p-values are below 0.05, the results are statistically significant. Overall, the variables show very strong associations. For instance, mpg exhibits a strong negative correlation with the other measures, suggesting that heavier and more powerful cars consume more fuel. Conversely, weight, displacement, and horsepower are strongly and positively correlated with one another, reflecting their close interdependence.

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization by origin
sns.boxplot(data=mpg,x='origin',y='displacement')
plt.show()
```

```python
# Assumption checks: Shapiro-Wilk + Levene's test
# origin sutununa gore displacement deyishenini qruplara ayırmaliyiq
usa = mpg[mpg['origin']=='usa']['displacement']
japan = mpg[mpg['origin']=='japan']['displacement']
europe = mpg[mpg['origin']=='europe']['displacement']

# Her qrup üçün Shapiro-Wilk morm paylanma var ya yox baxiriq

statistic_u, p_value_u = stats.shapiro(usa)
statistic_j, p_value_j = stats.shapiro(japan)
statistic_e, p_value_e = stats.shapiro(europe)

# Neticeleri çap edirik
print(f"Shapiro-Wilk Test:")
print(f"Statistic for usa: {statistic_u:.4f}")
print(f"P-value for usa: {p_value_u:.4f}")
print()
print(f"Statistic for japan: {statistic_j:.4f}")
print(f"P-value for japan: {p_value_j:.4f}")
print()
print(f"Statistic for europe: {statistic_e:.4f}")
print(f"P-value for europe: {p_value_e:.4f}")

# Levene testi qruplarin dispersiyalarinin birabir olub-olmadığini yoxlayir
levene_stat, levene_p = stats.levene(usa,japan,europe)

print(f"Levene's test: statistic={levene_stat:.4f}, p={levene_p:.4f}")
```

```python
# One-way ANOVA
# Bu test USA, Yaponiya və Evropa qruplari arasinda orta deyerlerin ferqli olub-olmadıgini yoxlayır
f_stat, p_value = stats.f_oneway(usa, japan, europe)

# F-statistika və p-value neticeleri chap olunur
print('F_stat: ', f_stat)
print('P_value: ',p_value)

# statsmodels kitabxanası anova model qurulur
import statsmodels.api as sm
from statsmodels.formula.api import ols

# mpg deyisheni origin qrupundan asili sekilde modelleşdirilir
model = ols('mpg ~ origin', data=mpg).fit()

# ANOVA table hesablanir
anova_table = sm.stats.anova_lm(model, typ=2)

# eta-squared hesablanir # *bilmedim buna bax****
esq = anova_table['sum_sq'].iloc[0] / (anova_table['sum_sq'].iloc[0] + anova_table['sum_sq'].iloc[1])

print(f'Eta-squared: {esq}')
```

```python
# Post-hoc test (Tukey's HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey = pairwise_tukeyhsd(endog=mpg['displacement'],  groups=mpg['origin'],  alpha=0.05)
print(tukey)
```

### Interpretation

*Which groups differ? What does the effect size tell you?*

Our statistical checks (Shapiro and Levene) confirmed that the data is non‑normal and the groups differ. We applied Tukey’s test, which revealed a clear gap: USA cars consume significantly more fuel compared to Japanese and European models. Japan and Europe show similar efficiency, both outperforming American cars in fuel economy. In short, Japanese and European vehicles stand together as far more fuel‑saving than their USA counterparts.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols].values
y = mpg["origin"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
# Train classifiers and generate predictions
# Lazim olan kitabxanalari import edirik
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Logistic Regression ucun melumatlari standartlasdiririq
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) 
 
# Random Forest ve Logistic Regression modellerini yaradirik
model_r = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42)
model_l = LogisticRegression(max_iter=1000)

# Modelleri train datasinda oyredirik
model_r.fit(X_train, y_train)
model_l.fit(X_train_scaled, y_train)

# Test datasinda proqnozlar aliriq
y_pred_r = model_r.predict(X_test)
y_pred_l = model_l.predict(X_test_scaled)
```

```python
# Classification reports
print('For RandomForestClassifier')
print(classification_report(y_test, y_pred_r))
print('For LogisticRegression')
print(classification_report(y_test, y_pred_l))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """Compute bootstrap 95% confidence interval for a classification metric.

    Returns
    -------
    point_estimate : float
    ci_lower : float
    ci_upper : float
    boot_scores : np.ndarray
    """

    # random eded generatoru yaradilir
    rng = np.random.RandomState(seed)

    # Numinelerin sayi
    n = len(y_true)

    # Bootstrap neticelerini saxlamaq ucun siyahi
    boot_scores = []

    # Original dataset ucun metrik deyerinin hesablanmasi
    point_estimate = metric_fn(y_true, y_pred)

    # Bootstrap yeniden numuneleme prosesi
    for _ in range(n_boot):

        # Tesadufi indeksler secilir (replacement ile)
        indices = rng.randint(0, n, size=n)

        # Eger secilmis numunede tek class varsa onu kecirik
        if len(np.unique(y_true[indices])) < 2:
            continue

        # Secilmis numune ucun metrik hesablanir
        score = metric_fn(y_true[indices], y_pred[indices])

        # Netice siyahiya elave olunur
        boot_scores.append(score)
 
    # Siyahi numpy arraye cevrilir
    boot_scores = np.array(boot_scores)

    # 95% confidence interval hesablanir
    ci_lower = np.percentile(boot_scores, 2.5)
    ci_upper = np.percentile(boot_scores, 97.5)

    # Neticeler qaytarilir
    return point_estimate, ci_lower, ci_upper, boot_scores
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)
# Random Forest modeli ucun bootstrap metrikleri hesablanir

# F1-score (weighted)
f1_point, f1_lo, f1_hi, f1_boots = bootstrap_metric(
    y_test,
    y_pred_r,
    lambda yt, yp: f1_score(yt, yp, average='weighted')
)

# Precision (weighted)
prec_point, prec_lo, prec_hi, prec_boots = bootstrap_metric(
    y_test,
    y_pred_r,
    lambda yt, yp: precision_score(yt, yp, average='weighted')
)

# Recall (weighted)
rec_point, rec_lo, rec_hi, rec_boots = bootstrap_metric(
    y_test,
    y_pred_r,
    lambda yt, yp: recall_score(yt, yp, average='weighted')
)
# Neticelerin cap edilmesi - Random Forest
print(f"Random Forest - F1: {f1_point:.4f} CI: [{f1_lo:.4f}, {f1_hi:.4f}]")
print(f"Random Forest - Precision: {prec_point:.4f} CI: [{prec_lo:.4f}, {prec_hi:.4f}]")
print(f"Random Forest - Recall: {rec_point:.4f} CI: [{rec_lo:.4f}, {rec_hi:.4f}]")

# Logistic Regression modeli ucun bootstrap metrikleri

# F1-score (weighted)
f1_point_l, f1_lo_l, f1_hi_l, f1_boots_l = bootstrap_metric(
    y_test,
    y_pred_l,
    lambda yt, yp: f1_score(yt, yp, average='weighted')
)

# Precision (weighted)
prec_point_l, prec_lo_l, prec_hi_l, prec_boots_l = bootstrap_metric(
    y_test,
    y_pred_l,
    lambda yt, yp: precision_score(yt, yp, average='weighted')
)

# Recall (weighted)
rec_point_l, rec_lo_l, rec_hi_l, rec_boots_l = bootstrap_metric(
    y_test,
    y_pred_l,
    lambda yt, yp: recall_score(yt, yp, average='weighted')
)

# Neticelerin cap edilmesi - Logistic Regression
print(f"Logistic Regression - F1: {f1_point_l:.4f} CI: [{f1_lo_l:.4f}, {f1_hi_l:.4f}]")
print(f"Logistic Regression - Precision: {prec_point_l:.4f} CI: [{prec_lo_l:.4f}, {prec_hi_l:.4f}]")
print(f"Logistic Regression - Recall: {rec_point_l:.4f} CI: [{rec_lo_l:.4f}, {rec_hi_l:.4f}]")
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)
# Metrikler
metrics = ['F1 Score', 'Precision', 'Recall']
#orta deyerleri
rf_means = [f1_point, prec_point, rec_point]
rf_errors = [ [f1_point - f1_lo, prec_point - prec_lo, rec_point - rec_lo],[f1_hi - f1_point, prec_hi - prec_point, rec_hi - rec_point]  ]

lr_means = [f1_point_l, prec_point_l, rec_point_l]
lr_errors = [
    [f1_point_l - f1_lo_l, prec_point_l - prec_lo_l, rec_point_l - rec_lo_l],
    [f1_hi_l - f1_point_l, prec_hi_l - prec_point_l, rec_hi_l - rec_point_l]  
]

x = np.arange(len(metrics))
width = 0.2
#qrafik olchusu
plt.figure(figsize=(10, 6))
#random forest ve logistic regression ucun errorbar grafik
plt.errorbar(x - width, rf_means, yerr=rf_errors, fmt='o', label='Random Forest', 
             capsize=5, markersize=8, elinewidth=2)
plt.errorbar(x + width, lr_means, yerr=lr_errors, fmt='o', label='Logistic Regression', 
             capsize=5, markersize=8, elinewidth=2)

plt.legend()
plt.show()
```

### Recommendation

*Which model would you recommend? How confident are you based on the CIs?*

The classification report indicates that Random Forest is the stronger model. It achieved an accuracy of 86%, compared to 81% for Logistic Regression. Bootstrap validation further confirms Random Forest’s advantage, showing higher F1, Precision, and Recall scores. Its 95% confidence intervals are also higher, underscoring greater stability and reliability for our dataset. In short, Random Forest is the more robust and dependable choice.
