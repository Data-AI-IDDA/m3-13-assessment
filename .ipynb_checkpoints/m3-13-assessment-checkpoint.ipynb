## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
# Correlation coefficients and p-values
num_cols = ["mpg", "horsepower", "weight", "acceleration"]
results = []

pairs = [
    ("mpg", "weight"),
    ("mpg", "horsepower"),
    ("horsepower", "weight")
]

for x, y in pairs:
    pearson_corr, pearson_p = pearsonr(mpg[x], mpg[y])
    spearman_corr, spearman_p = spearmanr(mpg[x], mpg[y])
    
    results.append({
        "Pair": f"{x} vs {y}",
        "Pearson r": round(pearson_corr, 3),
        "Pearson p": pearson_p,
        "Spearman r": round(spearman_corr, 3),
        "Spearman p": spearman_p
    })

results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Correlation heatmap
corr = mpg[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Scatter plots with regression lines
sns.lmplot(data=mpg, x="weight", y="mpg", height=5)
plt.title("Weight vs MPG")
plt.show()

sns.lmplot(data=mpg, x="horsepower", y="mpg", height=5)
plt.title("Horsepower vs MPG")
plt.show()
```

### Interpretation

*When is Pearson vs. Spearman more appropriate? Use your results as examples.*

Pearson and Spearman correlations are both used to measure relationships between variables, but they are appropriate in different situations. Pearson correlation measures linear relationships, while Spearman correlation measures monotonic relationships and is more robust to outliers.

For mpg vs weight, the Pearson correlation is -0.832 and the Spearman correlation is -0.876. These values are very close and strongly negative, and the scatter plot shows a clear linear downward trend. This indicates that the relationship is linear, so Pearson is appropriate.

For mpg vs horsepower, the Pearson correlation is -0.778 and the Spearman correlation is -0.854. Spearman is slightly stronger, and the scatter plot shows a negative but slightly curved relationship. This suggests the relationship is monotonic but not perfectly linear, so Spearman is more appropriate.

For horsepower vs weight, the Pearson correlation is 0.865 and the Spearman correlation is 0.879. Both values are very strong and similar, and the heatmap also confirms a strong positive relationship. This indicates a linear relationship, so Pearson is appropriate.

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization by origin
plt.figure(figsize=(8,5))
sns.boxplot(data=mpg, x="origin", y="mpg")
plt.title("MPG Distribution by Origin")
plt.show()

plt.figure(figsize=(8,5))
sns.violinplot(data=mpg, x="origin", y="mpg")
plt.title("MPG Distribution by Origin (Violin Plot)")
plt.show()
```

```python
# Assumption checks: Shapiro-Wilk + Levene's test
# Split groups
usa = mpg[mpg["origin"] == "usa"]["mpg"]
europe = mpg[mpg["origin"] == "europe"]["mpg"]
japan = mpg[mpg["origin"] == "japan"]["mpg"]

# Shapiro-Wilk test (normality)
print("Shapiro-Wilk Test:")
for name, group in zip(["USA", "Europe", "Japan"], [usa, europe, japan]):
    stat, p = shapiro(group)
    print(f"{name}: W = {stat:.4f}, p-value = {p:.4f}")

# Levene's test (equal variances)
stat, p = levene(usa, europe, japan)
print("\nLevene’s Test:")
print(f"Statistic = {stat:.4f}, p-value = {p:.4f}")
```

```python
# One-way ANOVA
f_stat, p_value = f_oneway(usa, europe, japan)

print("ANOVA Results:")
print(f"F-statistic = {f_stat:.4f}")
print(f"p-value = {p_value:.6f}")
```

```python
# Effect size (eta-squared)
# Overall mean
grand_mean = mpg["mpg"].mean()

# Between-group sum of squares
ss_between = (
    len(usa) * (usa.mean() - grand_mean)**2 +
    len(europe) * (europe.mean() - grand_mean)**2 +
    len(japan) * (japan.mean() - grand_mean)**2
)

# Total sum of squares
ss_total = ((mpg["mpg"] - grand_mean)**2).sum()
eta_squared = ss_between / ss_total
print(f"Eta-squared = {eta_squared:.4f}")
```

```python
# Post-hoc test (Tukey's HSD)
tukey = pairwise_tukeyhsd(
    endog=mpg["mpg"],
    groups=mpg["origin"],
    alpha=0.05
)

print(tukey)
```

### Interpretation

*Which groups differ? What does the effect size tell you?*

The ANOVA results show a statistically significant difference in MPG across the three regions (F = 96.60, p < 0.05), meaning that at least one group differs from the others.
The Tukey HSD post-hoc test reveals that all pairs of groups are significantly different. Specifically, Japanese cars have the highest MPG, followed by European cars, while USA cars have the lowest MPG. The differences between Europe vs Japan, Europe vs USA, and Japan vs USA are all statistically significant.
The effect size (η² = 0.3318) indicates a large effect, meaning that about 33% of the variation in MPG can be explained by the car’s origin. This suggests that origin has a strong practical impact on fuel efficiency, not just a statistically significant one.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols]
y = mpg["origin"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

I used .values before, but I changed it because it removes the DataFrame structure and index, which can cause misalignment issues during bootstrap resampling. I kept the data as pandas objects instead because it preserves indexing and makes the operations safer and more consistent.  
X = mpg[numeric_cols].values  
y = mpg["origin"].values to  

X = mpg[numeric_cols]  
y = mpg["origin"]

```python
# Train classifiers and generate predictions
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)

# Train
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Predictions
y_pred_dt = dt.predict(X_test)
y_pred_rf = rf.predict(X_test)
```

```python
# Classification reports
print("Decision Tree:\n")
print(classification_report(y_test, y_pred_dt))

print("\nRandom Forest:\n")
print(classification_report(y_test, y_pred_rf))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """Compute bootstrap 95% confidence interval for a classification metric.

    Returns
    -------
    point_estimate : float
    ci_lower : float
    ci_upper : float
    boot_scores : np.ndarray
    """
    # Convert to numpy arrays so positional indexing works
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    
    rng = np.random.default_rng(seed)
    n = len(y_true)
    scores = []

    for _ in range(n_boot):
        idx = rng.integers(0, n, n)   # resample with replacement
        score = metric_fn(y_true[idx], y_pred[idx])
        scores.append(score)

    scores = np.array(scores)
    point_estimate = scores.mean()
    ci_lower = np.percentile(scores, 2.5)
    ci_upper = np.percentile(scores, 97.5)

    return point_estimate, ci_lower, ci_upper, scores
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)
metrics = {
    "F1": lambda y_true, y_pred: f1_score(y_true, y_pred, average="macro"),
    "Precision": lambda y_true, y_pred: precision_score(y_true, y_pred, average="macro"),
    "Recall": lambda y_true, y_pred: recall_score(y_true, y_pred, average="macro"),
}

results = []

for model_name, y_pred in [("Decision Tree", y_pred_dt), ("Random Forest", y_pred_rf)]:
    for metric_name, metric_fn in metrics.items():
        point, low, high, _ = bootstrap_metric(y_test, y_pred, metric_fn)
        
        results.append({
            "Model": model_name,
            "Metric": metric_name,
            "Score": point,
            "CI Lower": low,
            "CI Upper": high
        })

import pandas as pd
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)
for metric in results_df["Metric"].unique():
    subset = results_df[results_df["Metric"] == metric]
    
    plt.figure()
    plt.errorbar(
        subset["Model"],
        subset["Score"],
        yerr=[subset["Score"] - subset["CI Lower"], subset["CI Upper"] - subset["Score"]],
        fmt='o',
        capsize=5
    )
    
    plt.title(f"{metric} with 95% CI")
    plt.ylabel(metric)
    plt.show()
```

### Recommendation

*Which model would you recommend? How confident are you based on the CIs?*

Based on the results, I would recommend the Decision Tree model. It shows better performance across all metrics compared to the Random Forest. Specifically, the Decision Tree has higher F1-score (0.83 vs 0.79), precision (0.85 vs 0.82), and recall (0.83 vs 0.77), as well as higher overall accuracy (0.89 vs 0.86).
The bootstrap confidence intervals also support this conclusion. For all metrics (F1, precision, and recall), the Decision Tree consistently has higher scores than the Random Forest. Although the confidence intervals overlap slightly, the Decision Tree’s intervals are generally shifted higher, indicating more reliable performance.
Therefore, I am fairly confident in recommending the Decision Tree model. The results suggest that it performs better not only on average but also across different resampled datasets, making it a more stable and suitable choice for this classification task.
