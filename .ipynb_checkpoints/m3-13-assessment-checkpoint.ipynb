```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

SEED = 42
np.random.seed(SEED)

mpg = sns.load_dataset("mpg").dropna()

print(f"Rows: {len(mpg)}, Columns: {mpg.shape[1]}")
print(f"Origin groups: {mpg['origin'].unique().tolist()}")
mpg.head()
```

## Task 1 — Correlation Analysis

1. Choose at least three pairs of numeric variables and compute Pearson and Spearman correlations.
2. Test each for significance (report p-values).
3. Create a correlation heatmap and at least two scatter plots with regression lines.
4. Explain when Pearson vs. Spearman is more appropriate, using your results as examples.

```python
# Correlation coefficients and p-values

pairs = [('mpg', 'weight'), ('mpg', 'horsepower'), ('horsepower', 'weight')]

for p1, p2 in pairs:
    pearson_r, p_val_p = stats.pearsonr(mpg[p1], mpg[p2])
    spearman_r, p_val_s = stats.spearmanr(mpg[p1], mpg[p2])
    
    print(f"--- {p1} vs {p2} ---")
    print(f"Pearson: r={pearson_r:.3f}, p-value={p_val_p:.3e}")
    print(f"Spearman: r={spearman_r:.3f}, p-value={p_val_s:.3e}\n")
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(mpg.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Numeric Variables Correlation Heatmap")
plt.show()

```

```python
# Scatter plots with regression lines
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Weight vs MPG 
r_weight, _ = stats.pearsonr(mpg['weight'], mpg['mpg'])
sns.regplot(data=mpg, x='weight', y='mpg', ax=axes[0], line_kws={'color':'red'})
axes[0].set_title(f"Weight vs MPG (r = {r_weight:.2f})")

# 2. Horsepower vs MPG 
r_hp, _ = stats.pearsonr(mpg['horsepower'], mpg['mpg'])
sns.regplot(data=mpg, x='horsepower', y='mpg', ax=axes[1], line_kws={'color':'red'})
axes[1].set_title(f"Horsepower vs MPG (r = {r_hp:.2f})")

plt.tight_layout()
plt.show()
```

### Interpretation

*When is Pearson vs. Spearman more appropriate? Use your results as examples.*

## Pearson vs. Spearman Correlation

**Pearson correlation** measures the **linear relationship** between variables.  
For example, the relationship between *weight* and *mpg* is largely linear, so Pearson is effective in this case.

**Spearman correlation**, on the other hand, measures a **monotonic relationship** (variables move in the same direction, but not necessarily in a straight line).  
It is also **more robust to outliers**.

For instance, the relationship between *horsepower* and *mpg* may be somewhat non-linear (curved), so Spearman can provide a more accurate result in this case.

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization 
plt.figure(figsize=(10, 6))
sns.boxplot(data=mpg, x='origin', y='mpg', hue='origin', palette='Set2', legend=False)
plt.title("MPG Distribution by Origin")
plt.show()


```

```python
# Assumption checks: Shapiro-Wilk + Levene's test

# Levene's Test (Equal Variances)
stat_lev, p_lev = stats.levene(
    mpg[mpg['origin'] == 'usa']['mpg'],
    mpg[mpg['origin'] == 'japan']['mpg'],
    mpg[mpg['origin'] == 'europe']['mpg']
)
print(f"\nLevene’s Test p-value: {p_lev:.4f}")


# Shapiro-Wilk (Normality)
print("Shapiro-Wilk Test (p-values):")
for group in mpg['origin'].unique():
    stat, p = stats.shapiro(mpg[mpg['origin'] == group]['mpg'])
    print(f"{group}: {p:.4f}")
```

```python
# One-way ANOVA
f_stat, p_val = stats.f_oneway(
    mpg[mpg['origin'] == 'usa']['mpg'],
    mpg[mpg['origin'] == 'japan']['mpg'],
    mpg[mpg['origin'] == 'europe']['mpg']
)

# Eta-squared calculation
def eta_squared(groups):
    combined = np.concatenate(groups)
    ss_total = np.sum((combined - np.mean(combined))**2)
    ss_between = sum(len(g) * (np.mean(g) - np.mean(combined))**2 for g in groups)
    return ss_between / ss_total

groups_data = [mpg[mpg['origin'] == g]['mpg'] for g in mpg['origin'].unique()]
es = eta_squared(groups_data)

print(f"ANOVA Results: F={f_stat:.2f}, p-value={p_val:.3e}")
print(f"Effect Size (Eta-squared): {es:.3f}")

```

```python
# Post-hoc test (Tukey's HSD)
# Tukey's HSD (if significant)
from statsmodels.stats.multicomp import pairwise_tukeyhsd
if p_val < 0.05:
    tukey = pairwise_tukeyhsd(endog=mpg['mpg'], groups=mpg['origin'], alpha=0.05)
    print("\nTukey HSD Test:")
    print(tukey)
```

### Interpretation

*Which groups differ? What does the effect size tell you?*

### Conclusion

- According to the ANOVA results, since the **p-value < 0.05**, there is a **statistically significant difference** in mpg (fuel efficiency) between the country of origin groups.  
- The **Tukey post-hoc test** shows that the **USA group** differs significantly from both the **Japan** and **Europe** groups. Based on this, **USA cars are less fuel-efficient**.  
- The calculated **eta-squared (η²)** indicates that the **country of origin explains a substantial portion of the variability in mpg**.

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols].values
y = mpg["origin"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
# Train classifiers and generate predictions
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Scaling for Logistic Regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
# Classification reports
# Models
rf_model = RandomForestClassifier(random_state=SEED)
lr_model = LogisticRegression(max_iter=1000, random_state=SEED)

rf_model.fit(X_train, y_train)
lr_model.fit(X_train_scaled, y_train)

y_pred_rf = rf_model.predict(X_test)
y_pred_lr = lr_model.predict(X_test_scaled)

print("Random Forest Report:\n", classification_report(y_test, y_pred_rf))
print("-" * 30)
print("Logistic Regression Report:\n", classification_report(y_test, y_pred_lr))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    np.random.seed(seed)
    boot_scores = []
    indices = np.arange(len(y_true))
    
    # Point estimate
    point_estimate = metric_fn(y_true, y_pred, average='weighted')
    
    for _ in range(n_boot):
        boot_idx = np.random.choice(indices, size=len(indices), replace=True)
        # Handle cases where bootstrap sample might missing a class
        score = metric_fn(y_true[boot_idx], y_pred[boot_idx], average='weighted')
        boot_scores.append(score)
        
    ci_lower = np.percentile(boot_scores, 2.5)
    ci_upper = np.percentile(boot_scores, 97.5)
    
    return point_estimate, ci_lower, ci_upper, np.array(boot_scores)
    pass
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)
metrics = {'F1': f1_score, 'Precision': precision_score, 'Recall': recall_score}
results = []

for name, preds in [('RF', y_pred_rf), ('LR', y_pred_lr)]:
    for m_name, m_fn in metrics.items():
        point, low, high, _ = bootstrap_metric(y_test, preds, m_fn)
        results.append({'Model': name, 'Metric': m_name, 'Score': point, 'Low': low, 'High': high})

res_df = pd.DataFrame(results)
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)
plt.figure(figsize=(10, 6))
for i, model in enumerate(['RF', 'LR']):
    subset = res_df[res_df['Model'] == model]
    plt.errorbar(x=subset['Metric'], y=subset['Score'], 
                 yerr=[subset['Score']-subset['Low'], subset['High']-subset['Score']], 
                 fmt='o', label=model, capsize=5)

plt.title("95% Bootstrap Confidence Intervals for Metrics")
plt.ylabel("Score")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Recommendation

*Which model would you recommend? How confident are you based on the CIs?*


Based on the conducted analysis and bootstrap results, I recommend using the **Random Forest (RF)** model for this task. The main reasons are as follows:

- **Higher Performance:** The Random Forest model slightly outperforms Logistic Regression in terms of both **F1-score** and **Precision**. This improvement comes from its ability to better capture **non-linear relationships** in the dataset (e.g., complex interactions between horsepower and origin).

- **Bootstrap Confidence Interval Analysis:** The **95% confidence intervals** indicate that the RF model has a more stable performance distribution. Although there is some overlap between the intervals of both models, the **mean performance of RF is consistently higher**.

- **Confidence and Robustness:** Since there is no large gap between the confidence intervals of the models, both can be considered reliable. However, the RF model demonstrates **greater robustness to outliers** and, due to its decision-tree-based structure, performs better in separating cars from different origins, which increases overall trust in its predictions.
